```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
profits = [168, 620, 1010]
actors = ["Emma Watson\n($168 million)", "Sebastian Stan\n($620 million)", "Mila Kunis\n($1.01 billion)"]

plt.pie(profits, labels=actors)
plt.axis("equal")
plt.title("How much profits hiring each actor brings for a $100 million movie", pad=32)
plt.show()
```
